```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
```

```python
client_credentials_manager = SpotifyClientCredentials(client_id=CLIENT_ID, client_secret=CLIENT_SECRET)
sp = spotipy.Spotify(client_credentials_manager = client_credentials_manager)
```

```python
def analyze_playlist(creator, playlist_id):
    
    # Create empty dataframe
    playlist_features_list = ["track_id", "track_name", "artist", "album", "acousticness",
                             "danceability", "energy", "key", "loudness", "mode", "speechiness",
                             "instrumentalness", "liveness", "valence", "tempo", "duration_ms", "time_signature"]
    playlist_df = pd.DataFrame(columns = playlist_features_list)
    
    # Create empty dict
    playlist_features = {}
    
    # Loop through every track in the playlist, extract features and append the features to the playlist df
    playlist = sp.user_playlist_tracks(creator, playlist_id)['items']
    for track in playlist:
        # Get metadata
        playlist_features["track_id"] = track["track"]["id"]
        playlist_features["track_name"] = track["track"]["name"]
        playlist_features["artist"] = track["track"]["album"]["artists"][0]["name"]
        playlist_features["album"] = track["track"]["album"]["name"]
        # Get audio features
        audio_features = sp.audio_features(playlist_features["track_id"])[0]
        for feature in playlist_features_list[4:]:
            playlist_features[feature] = audio_features[feature]
        
        # Concat the dfs
        track_df = pd.DataFrame(playlist_features, index = [0])
        playlist_df = pd.concat([playlist_df, track_df], ignore_index = True)
        
    return playlist_df
```

```python
playlist_df1 = analyze_playlist("Acclaimed Music", "6rU2YnPAA90aZfcS4rShB5")
playlist_df2 = analyze_playlist("Magnus Johansson", "1GJYdE69XSYadiI1TJseaa")
playlist_df3 = analyze_playlist("Acclaimed Music", "0nhHzY5aQZEATLUxbgOu5W")
playlist_df4 = analyze_playlist("Svavar Benediktsson", "5Eblumz1tZG317eDCKgZRQ")
playlist_df5 = analyze_playlist("rodrigotoledon", "73z25n97LBUfJj2QmiOvcw")
playlist_df6 = analyze_playlist("Jonathan Doughty", "1jZErtXIVlLcdtFrJvloDG")
playlist_df7 = analyze_playlist("Jeff Frankus", "6Bp4X2L9sFSu6yuCXwzkKJ")
playlist_df8 = analyze_playlist("Nick Tunes", "3ZnuEVhLfCqEInPqyrjNvK")
playlist_df9 = analyze_playlist("Daniel Grant", "6f20yOlE4LloK4COF3lS30")
playlist_df10 = analyze_playlist("egetunaveli", "6XNYd1GC5Ym73cn75HPiqR")
```

```python
playlists = [playlist_df1, playlist_df2, playlist_df3, playlist_df4, playlist_df5, playlist_df6, playlist_df7, playlist_df8, playlist_df9, playlist_df10]
playlist_df = pd.concat(playlists)
```
